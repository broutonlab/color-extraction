```python
import cv2
from matplotlib import pyplot as plt
import sys
sys.path.insert(0, '../')
from utils.topological_colors_extraction import extract_common_colors
from utils.visualization import add_color_bar_to_image

%matplotlib inline
```

```python
image_path = '../examples/car.jpg'
img = cv2.imread(image_path, cv2.IMREAD_COLOR)

if img is None:
    print('Can\'t load image: {}'.format(image_path))
    
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
common_colors, cover_rates = extract_common_colors(img)
```

```python
img_with_colors = add_color_bar_to_image(img, common_colors, cover_rates)

plt.figure(figsize=(10, 10))
plt.axis('off')
plt.imshow(img_with_colors)
plt.show()

# cv2.imwrite('../data/Lenna.png', cv2.cvtColor(img_with_colors, cv2.COLOR_RGB2BGR))
```
